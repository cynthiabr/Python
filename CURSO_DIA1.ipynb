```python
# Tipos de Datos
###Existen datos:
enteros = 4 #números sin decimales
flotantes = 3.9 #números con decimales
###Caracteres/string
caracteres = "a"
cadena_caracteres = "Hola" #todas las letras que quieras"
###Booleanas
falso = False #falso = 0
verdadero = True #verdadero = 1
###Imprimir
print(enteros)
###El usuario también puede ingresar cualquier tipo de dato
###y podemos guardarlo en alguna variable
name = input("Mi nombre es:") #te lo guarda como un string
print(name)
num = int(input("Ingresa un numero:"))
print(num + 1)
```

```python
#Operaciones Matemáticas
##4to Nivel de Importancia
###Suma ordinaria
print(3+4) #un 7
variable1, variable2 = 20,40
print(variable1+variable2) #un 60
###Resta ordinaria
print(2-5)
#Declaración de variables
v1= 10.4
v2 = 15
resultado = v2-v1
print(resultado)
```

```python
##3er Nivel de Importancia
###Multiplicación ordinaria
print("Multiplicación ordinaria:")
M1,M2=2,3.14
print(M1*M2)
###División ordinaria
print("División ordinaria:")
M1,M2,M3,M4,M5,M6 =2,5,6,5,4,5
print(M1/M2)
print(M3/M4)
print(M5/M6)
###División de Piso
###Redondea al entero abajo siempre
print("División de Piso:")
print(M1//M2)
print(M3//M4)
print(M5//M6)
```

```python
##2do Nivel de importancia
###Potencias
print("Potencia")
print(2**3) #8
P1, P2 = 7,2
print(P1**P2)
###Área de un círculo
print("Área de un círculo")
pi=3.14
r=4
A=pi*(r**2)
print(A)
```
