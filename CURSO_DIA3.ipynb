#Estructura de Datos
Las estructuras de datos son formas de organizar información para trabajar con ella de manera eficiente.
- Listas -> ordenadas, modificadas, permiten duplicados
- Tuplas -> ordenadas, inmutables, permiten duplicados
- Sets -> no ordenados, sin duplicados
- Diccionarios -> pares clave: valor

Notas: #investigar
- Pilas (FILO -> First in, Last out)
- Colas (FIFO -> First in, First out)
- Listas enlazadas (o doblemente enlazadas)

#Listas
Las listas son colecciones ordenadas que podemos modificar y almacenan elementos de distintos tipos de datos

```python
lista_de_compras = ["manzanas","pan","leche","huevos","mantequilla","queso","jamón"]
print(lista_de_compras)
print(lista_de_compras[6])
```

```python
#Agregar un elemento a la lista .append()
lista_de_compras.append("sopa")
print(lista_de_compras)
```

```python
#Agregar un elemento en una posición específica
lista_de_compras.insert(1,"plátano")
print(lista_de_compras)
```

```python
#Para eliminar un elemento de una lista existe el remove y el pop
lista_de_compras.remove("plátano") #quita el primero
print(lista_de_compras)
```

```python
lista_de_compras.pop #sin nada dentro del par+entesis borra el último elemento
lista_de_compras.pop(7) #o puedes agarrar la posició n que quieras pero iniciando desde 0
#Nota: te retorna el número eliminado
```

```python
#Podemos recorrer el arreglo por medio de un for
for elementos in lista_de_compras:
  print(elementos)
```

```python
#len() saca el tamaño de la lista
print(len(lista_de_compras))
for i in range(len(lista_de_compras)):
  print(f"{i+1}-{lista_de_compras[i]}")
```

```python
import random
opciones = ["piedra", "papel", "tijera"]

while True:
  while True:
    elección = input("\nElige piedra, papel o tijera (o escribe salir si ya no quieres jugar): ")
    if elección not in opciones and not elección == "salir":
      print("Opción no válida, intenta de nuevo")
    else:
      break

  if elección == "salir":
    print("¡Gracias por jugar!")
    break
  aleatoria = random.choice(opciones)
  print("La computadora eligió", aleatoria)

  #Determinamos el ganador
  if elección == aleatoria:
    print("Empate")
  elif (elección == "piedra" and aleatoria == "tijera") \
    or (elección =="papel" and aleatoria == "piedra") \
    or (elección == "tijera" and aleatoria == "papel"):
    print("¡Ganaste! :)!")
  else:
    print("Perdiste :(")
```

```python
promedios = [55.6, 97.5, 23.6, 97.5, 72.4, 100]
print(promedios)
promedios.sort() #sort sirve para ordenar la lista de manera ascendente
print(promedios)
#ordenar una lista de manera descendente
promedios.sort(reverse = True)
print(promedios)
```

```python
#podemos ingresar al último elemento de una lista
print(promedios[-1])
```

```python
#Encontrar el mayor y el menor de una lista
print(max(promedios))
print(min(promedios))
```

```python
#Declaración y usos de matricces
calificaciones = [[100,88,93],[99,56,72]]
i=1
for cali_i in calificaciones:
  print(f"\nAlumno{i}")
  j=1
  for cali_j in cali_i:
    print(f"Cali{j}: {cali_j}")
    j+=1
  j=0
  i+=1
```

#Sets
Los sets son colecciones SIN orden y que NO permiten elementos duplicados

```python
nombre = {"Ángel", "Eliud", "Diana", "Juan"}
print(nombre)
print(type(nombre))
```

```python
distintos_datos = {"Angel", 1, "Diana", "Juan", "Juan", 4.8}
print(distintos_datos)
```

```python
#Add
nombre.add("Leonardo")
print(nombre)
```

```python
print(nombre[2]) #marca error
nombre[2] #marca error
```

```python
#Union
matutino = {20322,22183,123123,53234}
vespertino = {20322,111111,222222,333333}
total_matrículas = matutino.union(vespertino) #union pero con .union()
print(total_matrículas)
total_matrículas =matutino | vespertino #union pero con |
print(total_matrículas)
```

```python
#Intersección
ca = {1234, 121, 646, 12414, 73434, 124124, 675234}
al = {1234, 111, 222, 333, 444}
#intersección con .intersection ()
ca_al= ca.intersection(al)
print(ca_al)
#intersección usando &
ca_al= ca & al
print(ca_al)
```

```python
#Diferencia
solo_cálculo = ca.difference(al) #Diferencia usando .difference
print(solo_cálculo)
#Diferencia usando -
solo_cálculo = ca - al
print(solo_cálculo)
```

```python
nombre = {"Angel", "Eliud", "Diana", "Juan", "Juan"}
#remove, pop
nombre.remove ("Angel") #Debes eliminar el que tu quieras
print(nombre)
nombre.pop() #Elimina un elemento de manera aleatoria
print(nombre)
```

```python
#Eliminar todo el SET
nombre.clear()
print(nombre)
```

#Diccionarios
Los diccionarios son colecciones no ordenadas de pares clave:valor

```python
alumno = {
    #clave     valor
    "nombre": "Cynthia Bolaños",
    "edad": 19,
    "carrera": "IB",
    "semestre": 3,
    "promedio": 93,
    "calificaciones": [100, 94, 94, 97, 100]
}
print(type(alumno))
print(alumno)
```

```python
print("El alumno", alumno["nombre"], "esta estudiando", alumno["carrera"])
```

```python
#Agregar al diccionario
alumno["area preferida"] = "Neuroingeniería"
print(alumno)
```
